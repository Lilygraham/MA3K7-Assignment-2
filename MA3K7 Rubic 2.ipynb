```python
import numpy as np
from scipy.linalg import det

def zerodet(matrix):
    return np.isclose(det(matrix), 0)

def generatematrix(n, numzero):
    matrix = np.random.choice([0, 1], size=(n, n), replace=True)
    
    zeroind = np.random.choice(range(n * n), size=numzero, replace=False)
    for index in zeroind:
        matrix[index // n][index % n] = 0
    
    return matrix

def calcprob(n, numzero, numtrial=10000):
    countzerodet = 0

    for _ in range(numtrial):
        matrix = generatematrix(n, numzero)
        if zerodet(matrix):
            countzerodet += 1

    prob = countzerodet / numtrial
    return prob

def main():
    n = int(input("Enter the dimension of the square matrix: "))
    numzero = int(input("Enter the number of 0 entries in the matrix: "))

    prob = calcprob(n, numzero)
    
    print(f"Estimate of the probability of obtaining a matrix with determinant 0:")
    print(f"For a {n}x{n} matrix with {numzero} zero entries: {prob:.4f}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def zerodet(matrix):
    return np.linalg.det(matrix) == 0

def calcprob(n, numtrial=1000):
    countzerodet = 0

    for _ in range(numtrial):
        matrix = np.random.randint(2, size=(n, n))
        if zerodet(matrix):
            countzerodet += 1

    prob = countzerodet / numtrial
    return prob

def generateprobs(maxn, numtrial=1000):
    probs = []

    for n in range(2, maxn + 1):
        prob = calcprob(n, numtrial)
        probs.append(prob)

    return probs

def plotprobs(probs):
    nvalues = list(range(2, len(probs) + 2))
    plt.plot(nvalues, probs, marker='o')
    plt.xlabel('Matrix Dimension ')
    plt.ylabel('Probability the Determinant of matrix is 0')
    plt.title('Estimated probability the Determinant of matrix is 0')
    plt.grid(True)
    plt.show()

def main():
    maxn = int(input("Enter the max value of n: "))
    numtrial = int(input("Enter the required number of samples for each n: "))

    probs = generateprobs(maxn, numtrial)
    plotprobs(probs)

if __name__ == "__main__":
    main()
```

```python
import itertools
import numpy as np

def zerodetmatrices(n, numzero):
    mats = []
    for combination in itertools.combinations(range(n*n), numzero):
        matrix = np.zeros((n, n), dtype=int)
        for idx in combination:
            row = idx // n
            col = idx % n
            matrix[row][col] = 1
        mats.append(matrix)
    return mats

def main():
    n = int(input("Enter the dimension of the square matrix: "))
    numzero = int(input("Enter the number of zero entries you need: "))

    mats = zerodetmatrices(n, numzero)

    print("The Matrices with determinant 0:")
    for matrix in mats:
        if np.linalg.det(matrix) == 0:
            print(matrix)


if __name__ == "__main__":
    main()
```
